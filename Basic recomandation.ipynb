```python
import pandas as pd
import numpy as np
```

```python
credit=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movie=pd.read_csv("tmdb_5000_movies.csv")
```

```python
credit_cloumn=credit.rename(index=str,columns={"movie_id":"id"})
movie_df_merge=movie.merge(credit_cloumn,on="id")
```

```python
movie_df_merge.head()
```

```python
for col in movie_df_merge.columns:
    print(col)
```

```python
movie_cleaned=movie_df_merge.drop(columns=["homepage","title_x","title_y","status","production_countries"])
```

```python
movie_cleaned.head()
```

```python
 movie_cleaned.info()
```

```python
v= movie_cleaned["vote_count"]
r= movie_cleaned["vote_average"]
c= movie_cleaned["vote_average"].mean()
m= movie_cleaned["vote_count"].quantile(0.70)
```

```python
movie_cleaned["weigth_average"]=((r*v)+(c*m))/(v+m)
```

```python
movie_cleaned.head()
```

```python
recomanded_ranking=movie_cleaned.sort_values("weigth_average",ascending=False)
recomanded_ranking[['original_title','vote_count','vote_average','weigth_average','popularity']].head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
weigth_average=recomanded_ranking.sort_values("weigth_average",ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weigth_average['weigth_average'].head(),y=weigth_average['original_title'].head())
plt.xlim(4,10)
plt.title('Best Movie by average votes',weight='bold')
plt.xlabel('Weigthed Average Score',weight='bold')
plt.ylabel('Movie title',weight='bold')
plt.savefig('best_movies.png')
```

# Based on popluarity 

```python
popularity_bsd =recomanded_ranking.sort_values("popularity",ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=popularity_bsd['popularity'].head(10),y=popularity_bsd['original_title'].head(10))
#plt.xlim(4,10)
plt.title('Best Movie by popularity',weight='bold')
plt.xlabel('popularity',weight='bold')
plt.ylabel('Movie title',weight='bold')
plt.savefig('best_movies.png')
```
